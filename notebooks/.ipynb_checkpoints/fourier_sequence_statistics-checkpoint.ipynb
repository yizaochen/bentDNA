### Part 1: Initialize

```python
hosts = ['a_tract_21mer', 'ctct_21mer', 'gcgc_21mer', 'g_tract_21mer', 'atat_21mer', 'tgtg_21mer']
group1 = ['a_tract_21mer', 'ctct_21mer', 'gcgc_21mer']
group2 = ['g_tract_21mer', 'atat_21mer', 'tgtg_21mer']
colors = ['b', 'g', 'r', 'c', 'm', 'y']
xticklabels = ['AA', 'CTCT', 'GCGC', 'GG', 'ATAT', 'TGTG']
```

```python
d_df = dict()
for host in hosts:
    s_agent = shapefourier.ShapeAgent(workfolder, host)
    d_df[host] = s_agent.read_df_an(n_begin, n_end)
```

### Part : Box Plot

```python
fig, axes = plt.subplots(ncols=5, nrows=2, figsize=(24,8))

n = 0
for row_id in range(2):
    for col_id in range(5):
        if (row_id == 1) and (col_id == 4):
            ax = axes[row_id, col_id]
            ax.set_axis_off()
            continue
        ax = axes[row_id, col_id]
        key = str(n)
        x = list()
        for host in hosts:
            x.append(d_df[host][key])
        ax.boxplot(x)
        ax.set_xticklabels(xticklabels)
        ax.set_ylabel(r'$a_{' + f'{n}' + r'}$', fontsize=14)
        ax.tick_params(axis='x', labelsize=12)
        n += 1
plt.tight_layout()
#plt.savefig("an_boxplot.png", dpi=200)
plt.show()
```
